#### 2. Análise para uma variável qualitativa

#### 2.1 Tabela de Frequências

```python
tabela_sexo = questionario['Sexo'].value_counts()

tabela_sexo
```

```python
tabela_sexo_porcentagem = questionario['Sexo'].value_counts()/len(questionario)*100

tabela_sexo_porcentagem
```

#### 2.2 Gráfico de barras verticais

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(tabela_sexo.index, tabela_sexo.values, color = "pink")
plt.title("Distribuição dos alunos por sexo")
plt.xlabel("Sexo")
plt.ylabel("Número de pessoas")

for i, v in enumerate (tabela_sexo.values):
        plt.text(i, v + 0.2, str(v), ha = "left", va = "center")
```

#### 2.3 Gráfico de barras horizontais

```python
plt.barh(tabela_sexo.index, tabela_sexo.values, color = "pink")
plt.title("Distribuição dos alunos por sexo")
plt.ylabel("Sexo")
plt.xlabel("Número de pessoas")

for i, v in enumerate (tabela_sexo.values):
        plt.text(v + 0.2, i, str(v), ha = "left", va = "center")
```
