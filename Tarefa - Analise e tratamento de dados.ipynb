```python
## - Implemente três soluções para tratar os dados faltantes: substituição pela média, substituição pela moda e remoção da linha.

# 1) Preenchendo os NaNs com a média
media_Ttrab = df['Ttrab'].mean()
df['Ttrab'].fillna(media_Ttrab, inplace = True)

media_Itrab = df['Itrab'].mean()
df['Itrab'].fillna(media_Itrab, inplace = True)

media_Serief = df['Serief'].mean()
df['Serief'].fillna(media_Serief, inplace = True)

df.isna().sum()
```

```python
## - Implemente três soluções para tratar os dados faltantes: substituição pela média, substituição pela moda e remoção da linha.

# 2) Preenchendo os NaNs com a moda
df = pd.read_excel("C:/Users/20221enpro0081/Downloads/aeusp.xlsx")

moda_Itrab = df['Itrab'].mode()[0] # [0] pega a primeira moda
df['Itrab'].fillna(moda_Itrab, inplace = True)

moda_Ttrab = df['Ttrab'].mode()[0] # [0] pega a primeira moda
df['Ttrab'].fillna(moda_Ttrab, inplace = True)

moda_Serief = df['Serief'].mode()[0] # [0] pega a primeira moda
df['Serief'].fillna(moda_Serief, inplace = True)

df.isna().sum()
```

```python
# - Baseado nas variáveis Sexo e ITrab, você diria que os homens começam a trabalhar mais cedo?

df.groupby(['Sexo'])['Itrab'].mean()

# 1: masculino
# 2: feminino
# Considerando a média de idades que as pessoas começaram a trabalhar, podemos dizer que os homens começam a trabalhare mais cedo que as mulheres
```
