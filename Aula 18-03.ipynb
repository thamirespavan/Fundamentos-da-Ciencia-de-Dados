```python
# Importando o conjunto de dados 

import pandas as pd

dados = pd.read_csv("C:/Users/20221enpro0081/Downloads/Salary_Data.csv")

dados
```

```python
# Diagrama de dispersão 

import matplotlib.pyplot as plt

plt.scatter(dados['YearsExperience'], dados['Salary'], color = 'hotpink')
plt.title("Salário vs Experiência")
plt.xlabel("Anos de experiência")
plt.ylabel("Salário (Dólares americanos por ano)")
```

```python
# Separando X e Y

# : > pega todas as linhas
# 0:5 pega da linha 0 até a 5, sem pegar a 5 
# :1 > pega tudo que vem antes da coluna 1 sem contar a 1

x = dados.iloc[:, :1]
y = dados.iloc[:, 1:]
```

```python
x
```

```python
y
```

```python
# Modelo

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

modelo = LinearRegression()
modelo.fit(x,y)
```

```python
# Coeficientes do modelo (betas)

print('Coeficiente Angular (beta_1): ', modelo.coef_)
print('Intercepto (beta_2):', modelo.intercept_)
```

```python
# Gerando predições

y_pred = modelo.predict(x)
```

```python
plt.scatter(x, y, color = "hotpink")
plt.plot(x, y_pred, color = "pink")
plt.title("Salário vs Experiência")
plt.xlabel("Anos de experiência")
plt.ylabel("Salário anual em dólares americanos")
```

```python
# Coeficiente de determinação (R²)
r2_score(y, y_pred)
```
